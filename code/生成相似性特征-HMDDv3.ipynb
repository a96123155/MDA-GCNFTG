```python
mirna_gip_similarity = np.zeros((len(mirna_idx), len(mirna_idx)))
mirna_gip_similarity.shape
```

```python
theta_r = 0
for i in trange(len(mirna_idx)):
    theta_r += np.linalg.norm(mirna_interaction_profile.iloc[i, 1:]) ** 2
theta_r = n / theta_r # n = len(mirna_idx)
print('theta_r = {}'.format(theta_r))

for i in trange(len(mirna_idx)):
    for j in range(len(mirna_idx)):
        if i == j: 
            mirna_gip_similarity[i,j] = 1
        else:
            temp = mirna_interaction_profile.iloc[i, 1:] - mirna_interaction_profile.iloc[j, 1:]
            temp = np.linalg.norm(temp) ** 2
            temp = - theta_r * temp
            temp = np.exp(temp)
            mirna_gip_similarity[i,j] = mirna_gip_similarity[j,i] = temp
mirna_gip_similarity
```

# disease semantic similarity

```python
D1 = np.identity(len(disease_name))
print(D1.shape)

def calc_D1(disease_idx_mesh, D1, di, dk):
    
        if D1[di, dk] != 0:
            return D1[di, dk]
        else:
            dk_children = disease_idx_mesh.loc[dk, 'children_idx']
            if dk_children == 0:
                return D1[di, dk] # 0
            else:
                temp = [0.5 * calc_D1(disease_idx_mesh, D1, di, dk_child) for dk_child in dk_children]
                D1[di, dk] = max(temp)
                if D1[di, dk] != 0 and D1[di, dk] != 0.5: print('di = {} | dk = {} | D1[di, dk] = {} | temp = {}'.format(di, dk, D1[di, dk], temp))
                return D1[di, dk]
                
for di in tqdm(disease_idx_with_mesh_id):
    for dk in disease_idx_with_mesh_id:
        if di != dk:
            D1[di, dk] = calc_D1(disease_idx_mesh, D1, di, dk)
#         break
D1
```

```python
DS1 = np.ones(len(disease_name)) # self node di
print(DS1.shape)

for di in tqdm(disease_idx_with_mesh_id):
    Tdi = disease_idx_mesh.loc[di, 'parent_idx']
    print('di = {} | Tdi = {}'.format(di, Tdi))
    
    if Tdi != 0:
        for Tdi_ in Tdi:
            DS1[di] += D1[di, Tdi_]
DS1
```

```python
DSSM1 = np.zeros((len(disease_name), len(disease_name)))
print(DSSM1.shape)

for di in tqdm(disease_idx_with_mesh_id):
    for dj in disease_idx_with_mesh_id:
        
        # Tdi_Tdj_intersection
        Tdi, Tdj = [di], [dj]

        if disease_idx_mesh.loc[di, 'parent_idx'] != 0:
            Tdi.extend(disease_idx_mesh.loc[di, 'parent_idx'])

        if disease_idx_mesh.loc[dj, 'parent_idx'] != 0:
            Tdj.extend(disease_idx_mesh.loc[dj, 'parent_idx'])
        Tdi_Tdj_intersection = list(set(Tdi).intersection(set(Tdj)))
#         print(Tdi, Tdj, Tdi_Tdj_intersection)
        
        if Tdi_Tdj_intersection == []:
            DSSM1[di, dj] = 0
        else:
            # numerator 分子
            numerator = 0
            for dt in Tdi_Tdj_intersection:
                temp = D1[di, dt] + D1[dj, dt]
                numerator += temp

            # denominator 分母
            denominator = DS1[di] + DS1[dj]

            # DSSM1[di, dj]
            DSSM1[di, dj] = numerator / denominator
            
            if numerator != 0:
                print(Tdi_Tdj_intersection, numerator, denominator, DSSM1[di, dj])
DSSM1
```

```python
D2 = np.zeros((len(disease_name), len(disease_name)))
print(D2.shape)

for di in tqdm(disease_idx_with_mesh_id):
    for dk in disease_idx_with_mesh_id:
        numerator = len(disease_idx_mesh.loc[dk, 'mesh_id'])
        denominator = mesh_id.shape[0]
        temp = numerator / denominator
        
        D2[di, dk] = - np.log(temp)
        
        if di == 5: print('di = {}, dk = {}, numerator = {}, denominator = {}, D2[di, dk] = {}'.format(di, dk, numerator, denominator, D2[di, dk]))
D2
```

```python
DS2 = np.ones(len(disease_name)) # self node di
print(DS2.shape)

for di in tqdm(disease_idx_with_mesh_id):
    Tdi = [di]
    if disease_idx_mesh.loc[di, 'parent_idx'] != 0:
        Tdi.extend(disease_idx_mesh.loc[di, 'parent_idx'])
        
    print('di = {} | Tdi = {}'.format(di, Tdi))

    for Tdi_ in Tdi:
        DS2[di] += D2[di, Tdi_]
DS2
```

```python
DSSM2 = np.zeros((len(disease_name), len(disease_name)))
print(DSSM2.shape)

for di in tqdm(disease_idx_with_mesh_id):
    for dj in disease_idx_with_mesh_id:
        
        # Tdi_Tdj_intersection
        Tdi, Tdj = [di], [dj]

        if disease_idx_mesh.loc[di, 'parent_idx'] != 0:
            Tdi.extend(disease_idx_mesh.loc[di, 'parent_idx'])

        if disease_idx_mesh.loc[dj, 'parent_idx'] != 0:
            Tdj.extend(disease_idx_mesh.loc[dj, 'parent_idx'])
        Tdi_Tdj_intersection = list(set(Tdi).intersection(set(Tdj)))
#         print(Tdi, Tdj, Tdi_Tdj_intersection)
        
        if Tdi_Tdj_intersection == []:
            DSSM2[di, dj] = 0
        else:
            # numerator 分子
            numerator = 0
            for dt in Tdi_Tdj_intersection:
                temp = D2[di, dt] + D2[dj, dt]
                numerator += temp

            # denominator 分母
            denominator = DS2[di] + DS2[dj]

            # DSSM2[di, dj]
            DSSM2[di, dj] = numerator / denominator
            
            if numerator != 0:
                print(Tdi_Tdj_intersection, numerator, denominator, DSSM2[di, dj])
DSSM2
```

```python
DSSM = (DSSM1 + DSSM2)/2
disease_semantic_similarity = pd.DataFrame(DSSM)
disease_semantic_similarity.drop(index = 743, columns = 743, inplace = True)
disease_semantic_similarity
```
